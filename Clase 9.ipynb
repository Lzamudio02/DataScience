# Probabilidad y estadística
Permiten estudiar fenómenos aleatorios para poder generar predecciones en su comportamiento.

### Distribuciones 

```python
#Un ejemplo de una distribución
mu= 107
sigma= 3
n=10000
x=np.random.normal(mu,sigma,n)
fig,ax= plt.subplots(1,1,figsize=(5,5))
sns.histplot(x,kde=True)
plt.title("Distribución normal")
```

### Media


mediana: dato de justo el medio, si los datos son impares se toman los dos del medio y se promedian. 


### Desviación estándar

```python
x.std()
```

### Varianza

```python
x.var()
```

### Curtosis

Indica la forma de la curva, el pico o la cuadratura. 


```python
media = x.mean()
n = len(x)
num=(((x-media)**4).sum())/n
den=((((x-media)**2).sum())/n)**2
num/den -3
```

```python
#Si se compara con la función de scipy 
scipy.stats.kurtosis(x) #Bastante similar
```

### Asimetría

```python
scipy.stats.skew(x)
```
