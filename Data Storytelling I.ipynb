```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data['Year'].unique()
```

```python
switzerland_df = data[data['Country or Area'] == 'Switzerland']
```

```python
plt.plot(switzerland_df['Year'], switzerland_df['Value'])
```

```python
countries = ['Ireland', 'Norway', 'Switzerland', 'Cayman Islands']
```

```python
fig, ax = plt.subplots(1, figsize=(6,6))

for country in countries:
    country_df = data[data['Country or Area'] == country]
    plt.plot(country_df['Year'], country_df['Value'])
    
    plt.text(country_df['Year'].values[0] + 0.02,country_df['Value'].values[0],
             country_df['Value'].iloc[0])
    plt.text(country_df['Year'].values[1] - 0.02,country_df['Value'].values[1], 
             country, ha='right')
    
    
plt.xlim(2017.5,2019.5)
plt.xticks([2018, 2019])

ax.xaxis.grid()

ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
```
