```python
def clean(X):
    # Imputar columnas numéricas
    X['GarageYrBlt'] = X['GarageYrBlt'].fillna(0)
    X['LotFrontage'] = X['LotFrontage'].fillna(0)
    X['MasVnrArea'] = X['MasVnrArea'].fillna(X['MasVnrArea'].mean())
    
    X['MSSubClass'] = X['MSSubClass'].astype(str)
    X['MoSold'] = X['MoSold'].astype(str)
    
    X.drop(columns = ["PoolQC", "MiscFeature","Alley","Fence","FireplaceQu","Id"], inplace=True)
    
    for col in cat_cols:
        X[col] = X[col].fillna(X[col].mode()[0])
        
def get_encoder(X):
    
    types = data.dtypes
    cat_cols = types[types == 'object'].index
    encoder = {}
    for col in cat_cols:
        encoder[col] = data.groupby(col)['SalePrice'].mean().to_dict()
    return encoder


def encode(X, encoder):
    for feature in encoder.keys():
        X[feature] = X[feature].map(encoder[feature])
        X[feature] = X[feature].fillna(0)
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
clean(train)
clean(test)
```

```python
encoder = get_encoder(train)
```

```python
encode(train, encoder)
encode(test, encoder)
```
