```python
import pandas as pd
import numpy as np
```

```python
datos = {
    'Presio': [100, 23, np.nan, np.nan, 43, 10, 23],
    'Kantidad': [9, 2, 2, np.nan, 8, 9, 4],
    'Productos Bendidos': [2, 45, 21, np.nan, 5, 20, np.nan]
}
```

### Renombrar columnas

```python
data = pd.DataFrame(datos)
data
```

```python
data.rename(columns={"Presio":"Precio",
                     "Kantidad":"Cantidad",
                    "Productos Bendidos":"Productos_Vendidos"},
           inplace=True)
```
