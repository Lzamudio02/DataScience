```python
#se creó un yaml para proteger las credenciales 
with open('config.yaml') as f:
    config = yaml.load(f, Loader=SafeLoader)
```

```python
#Pasamos las credenciales del yaml
user = config['user']
passw = config['passw']
host = config['host'] #direccion del servido url o ip
db_name = config['db_name']
port = config['port']
```

```python
#el %s significa que debe ser sustituido por otro valor 
connect_str = "mysql://%s:%s@%s:%s/%s?charset=utf8" %(user, passw,host, port, db_name)
```

```python
engine = create_engine(connect_str)
```
