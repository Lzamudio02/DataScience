# KNN vecinos cercanos


```python
import pandas as pd
```

```python
iris = pd.read_json("C:/Users/USA/Downloads/iris.json")
```

```python
iris
```

```python
from sklearn import preprocessing
```

```python
#Tomo la variable que quiere predecir
y = iris['species']
```

```python
#Ahora debemos codificar 
le = preprocessing.LabelEncoder()
le
```

```python
le.fit(y)
```

```python
y = le.transform(y)
```

```python
y
```

```python
X = iris[['sepalLength','sepalWidth','petalLength','petalWidth']]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
#A este método se le pasa las variables predictoras y la variable respuesta
model.fit(X, y)
```

```python
model.predict(X)
```

```python
#Vector de probabilidades
model.predict_proba(X)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X, y)
```

```python
model.predict(X)
```
