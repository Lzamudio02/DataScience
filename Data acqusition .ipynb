# API

```python
import requests
```

```python
endpoint = 'https://pokeapi.co/api/v2/pokemon/'
```

```python
r = requests.get(endpoint)
```

```python
r.status_code
```

```python
data = r.json()
```

```python
data
```

```python
data['results']
```

```python
pokemons = []
names=[]
for pokemon in data['results']:
    name = pokemon["name"]
    url_pokemon = pokemon['url']
    r = requests.get(url_pokemon)
    
    if 200 != r.status_code:
        continue
    
    pokemon_data = r.json()
    pokemons.append(pokemon_data)
    names.append(name)
```

```python
pokemons[1]
```

```python
data.keys()
```

```python
df=pd.json_normalize(pokemons)
df
```

```python
df["names"]=names
```

```python
df.set_index("names",inplace=True)
```
