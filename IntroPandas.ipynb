Una serie es diferente a una lista, recibe una lista como argumento.
Para iterar una serie es igual a una lista.
Una serie me permite elegir los índices que yo quiera.

```python
lista = list(np.arange(0,10))

serie = pd.Series(["Za",3,4,5],index=[1,"casa",5,'MARIA'])
print(serie)
```

```python
#Operaciones entre series

serie*2
```

```python
#Acceder a elementos
serie[1]
```

```python
serie[1:] #recorre desde la segunda posición hasta la última
```

```python
serie[:2] #recorre desde la primera posición hasta la tercera posición
```

# Moverse en dataframes 

```python
ruta="C:/Users/USA/Downloads/winequality-red.csv"
archivo=pd.read_csv(ruta) #CSV

#Columnas:
archivo[["alcohol","quality"]]#Accede a las dos columnas llamadas xoy
```

```python
#Filas
"""Si quiero ver el dato de la fila 4 de la columna llamada alcohol uso LOC"""
archivo['alcohol'].loc[4]
```

```python

"""Los nombres de las columnas aparecen a la izq y el 
dato toda la fila 4 a parecen junto a el nombre"""
archivo.loc[2] 
```

inplace=True es muy importante pues al hacer ciertas operaciones como añadir o remover datos por lo general no se guarda como una nueva variable. Al hacer inplace=True, estas operaciones se irán guardando en el nuevo dataframe sin tener que declarar una nueva variable.

```python

#Eliminar columnas
archivo.drop(columns="alcohol")
```

```python
archivo #Al imprimir otra vez el frame aún aparece alcohol
        #porque no se uso el inplace
```

```python
archivo.drop(2) #elimina la fila 2
```

# Filtros / operaciones lógicas

```python
#Elegir los vinos con azúcar < 2.2
archivo["residual sugar"]<2.2
```

```python
archivo[archivo["residual sugar"]<1.5]
```

```python
a=archivo[archivo["residual sugar"]<1.5]
print("Solo {} vinos cumplen este requisito.".format(len(a)))
```

```python
archivo[(archivo["pH"]>3.49)&(archivo["quality"]>5)]
#Me elige los vino con calidad > 5 y con pH > 3.49
```

```python
#Cambiar un valor
#Cambiar el valor de la fila 4 col alcohol a 11

archivo.loc[4,"alcohol"]=11
archivo
```
